```python
import json

# Ler como JSON
with open('/content/drive/MyDrive/Dados Telecom 1/TelecomX_Data.json', 'r') as f:
    dados_json = json.load(f)

# Normalizar o df
df_normalizado = pd.json_normalize(dados_json, sep='_')
df_normalizado
```

```python
for col in df_normalizado.columns:
  unicos = df_normalizado[col].unique()
  print(f"---Coluna '{col}': ---")
  print(f"valores Ãºnicos -> {unicos}")
  print('\n')
```

```python
#Detectar valores vazios '' na coluna Churn
print((df_normalizado['Churn'] == '').sum())
```

```python
df_normalizado = df_normalizado.replace({'': pd.NA})
print(df_normalizado['Churn'].isna().sum())
```

```python
print('Linhas vazias no Churn:', (df_normalizado['Churn'] == '').sum())
```

```python
#Detectar valores na
print(df_normalizado.isna().any())
```

```python
#Encontrar valores duplicados
duplicados = df_normalizado[df_normalizado.duplicated()]
duplicados
```

```python
#Ver os tipos de dados de cada coluna
df_normalizado.info()
```

```python
print((df_normalizado['account_Charges_Total'] == " ").sum())
```

```python
#Modificar tipo de dado da coluna account_Charges_Total
df_normalizado['account_Charges_Total'] = pd.to_numeric(df_normalizado['account_Charges_Total'].astype(str).str.strip(),
                                                        errors = 'coerce')
```

```python
#Modificar tipo de dado da coluna account_Charges_Total
df_normalizado['account_Charges_Total'] = df_normalizado['account_Charges_Total'].astype(np.float64)
```

```python
#Padronizar os nomes das colunas do df
import re

def adicionar_underscore(nome_coluna):
    # Insere _ antes de maiÃºsculas e coloca tudo em minÃºsculas
    nome = re.sub(r'([a-z])([A-Z])', r'\1_\2', str(nome_coluna)).lower()
    return nome.strip('_')

# Aplica em todas as colunas
df_normalizado.columns = [adicionar_underscore(col) for col in df_normalizado.columns]

df_normalizado
```

```python
from plotly.subplots import make_subplots

payment_order = [
    'Electronic check',
    'Mailed check',
    'Bank transfer (automatic)',
    'Credit card (automatic)'
]


# Criar subplots grÃ¡ficos de pizza
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "pie"}, {"type": "pie"}]],
    subplot_titles=('No', 'Yes'),
    horizontal_spacing=0.40
)

# 1. Filtrar dados Churn = No
df_no = df_normalizado[df_normalizado['churn'] == 'No']['account_payment_method'].value_counts().reindex(payment_order,fill_value=0)

# 2. Filtrar dados Churn = Yes
df_yes = df_normalizado[df_normalizado['churn'] == 'Yes']['account_payment_method'].value_counts().reindex(payment_order,fill_value=0)

# 3. Adicionar grÃ¡ficos de pizza
fig.add_trace(
    go.Pie(labels=df_no.index, values=df_no.values, name="Churn=No", hole = 0.6, sort=False, direction='clockwise',
        rotation=0, showlegend=False,
           textposition='outside', textinfo='percent+label', pull = [0, 0.18,0,0]),
           row=1, col=1
)

fig.add_trace(
    go.Pie(labels=df_yes.index, values=df_yes.values, name="Churn=Yes", hole = 0.6, sort=False, direction='clockwise',
        showlegend=False, textposition='outside', textinfo='percent+label', pull = [0.15, 0,0,0]),
    row=1, col=2
)

# 4. Layout e exibir
fig.update_layout(height=500, width=1000, title_text="MÃ©todos de Pagamento por Churn")
fig.show()
```

##**IntroduÃ§Ã£o**

Baseada em um cenÃ¡rio de elevado Ã­ndice de cancelamentos, esta anÃ¡lise teve como objetivo identificar os principais fatores associados Ã  perda de clientes (churn).
Para isso, foram realizadas etapas de ExtraÃ§Ã£o, TransformaÃ§Ã£o e Carga (ETL) dos dados e, em seguida, o churn foi analisado com base em variÃ¡veis categÃ³ricas e numÃ©ricas, cujos resultados sÃ£o detalhados nas seÃ§Ãµes seguintes deste relatÃ³rio.


#**ExtraÃ§Ã£o dos dados:**

- A extraÃ§Ã£o de dados foi realizada com o uso da biblioteca Pandas. O arquivo JSON, armazenado no Google Drive, foi inicialmente lido por meio do mÃ©todo pd.read_json, tendo o resultado atribuÃ­do Ã  variÃ¡vel dados.

- A inspeÃ§Ã£o inicial com o mÃ©todo .head() evidenciou que os registros estavam aninhados, o que motivou a uma nova leitura do arquivo Telecom_X em modo leitura 'r' com o mÃ³dulo json. Nessa etapa, o conteÃºdo foi convertido em uma lista de dicionÃ¡rios por meio de jsaplicou-se pd.json_normalize, para transformar essa lista em um DataFrame (df_normalizado), utilizando o caractere '_' como separador na composiÃ§Ã£o dos nomes das colunas.

#**Limpeza e Tratamento de Dados:**

- A etapa de limpeza e tratamento teve inÃ­cio com a inspeÃ§Ã£o dos valores Ãºnicos em cada coluna do df_normalizado, por meio de um loop, o que permitiu verificar a presenÃ§a de valores vazios (' ') na coluna churn. Esses valores foram substituÃ­dos por NA.


- Adicionalmente, verificou-se que o DataFrame nÃ£o possuÃ­a valores nulos ou duplicados.


- Durante a anÃ¡lise dos tipos de dados, observou-se por meio de .info() que a coluna account_Charges_Total encontrava-se armazenada como object, embora devesse ser numÃ©rica(float). A coluna foi tratada como texto e em seguida convertida ao tipo numÃ©rico. Por fim a coluna foi convertida para float64 com `astype('float64').`


- Por Ãºltimo, foi definida uma funÃ§Ã£o utilizando o mÃ³dulo re com o objetivo de padronizar os nomes das colunas. Essa funÃ§Ã£o inseriu um underscore(_) antes de letras maiÃºsculas e converteu todo o texto para minÃºsculas, facilitando a leitura e manuseio das variÃ¡veis ao longo da anÃ¡lise.
